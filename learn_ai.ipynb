```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()

print(diabetes.data.shape, diabetes.target.shape)
print(diabetes.data[0:3])
print(diabetes.target[:3])
```

```python
plt.scatter(diabetes.data[:, 2], diabetes.target)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# gradient descent
x = diabetes.data[:, 2]
y = diabetes.target

w = 1.0
b = 1.0

y_hat = x[0] * w + b
print('y:', y[0], '\n' + 'y_hat:', y_hat)

w_inc = w + 0.1
y_hat_inc = x[0] * w_inc + b
print('y_hat_inc:', y_hat_inc)

w_rate = (y_hat_inc - y_hat) / (w_inc - w)
print('w_rate:', w_rate)

w_new = w + w_rate
print('w_new:', w_new)

b_inc = b + 0.1
y_hat_inc = x[0] * w + b_inc
print('y_hat_inc:', y_hat_inc)

b_rate = (y_hat_inc - y_hat) / (b_inc - b)
print('b_rate:', b_rate)

b_new = b + 1
print('b_new:', b_new)

# 단점 y_hat이 y에 한참 미치지 못하는 경우, w와 b를 더 큰 폭으로 수정할 수 없음.(위에서는 변화율 만큼 수정을 했지만 특별한 기준을 정할 수 없음)
#      y_hat이 y보다 커지면 y_hat을 감소시키지 못함
```
